```python
# Import dependency
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
```

```python
# create engine and pass through postgress connection
engine = create_engine('postgresql://username:password@localhost:5432/Employee_db')
connection=engine.connect()
```

```python
#Query records in the departments table
departments_df=pd.read_sql("select * from departments", connection)
departments_df.head()
```

```python
#Query records in the dept_emp table
dept_emp_df=pd.read_sql("select * from dept_emp", connection)
dept_emp_df.head()
```

```python
#Query records in the dept_manager table
dept_manager_df=pd.read_sql("select * from dept_manager", connection)
dept_manager_df.head()
```

```python
#Query records in the employees table
employees_df=pd.read_sql("select * from employees", connection)
employees_df.head()
```

#Read the databases for titles and salaries into dataframes

```python
#Query records in the salaries table
salaries_df=pd.read_sql("select * from salaries", connection)
salaries_df.head()
```

```python
#Query records in the titles table
titles_df=pd.read_sql("select * from titles", connection)
titles_df
```

```python
#merge employees and titles dataframe
combined_df=pd.merge(employees_df, titles_df, left_on='emp_title_id',right_on='title_id')
combined_df.head()
```

```python
#merge combined_df and salary dataframe
combined_df1=pd.merge(combined_df, salaries_df, on ="emp_no", how ="inner")
combined_df1.head()
```

```python
#group by with titles
combined_grp=combined_df1.groupby("title").mean()
combined_grp
```

```python
#drop column emp_no
combined_grp_df=combined_grp.drop(columns = "emp_no")
combined_grp_df
```

```python
#put index number
combined_grp_df=combined_grp_df.reset_index()
combined_grp_df
```

```python
#create histogram
#plt.style.use(ggplot)
x_axis=combined_grp_df["title"]
y_axis=combined_grp_df["salary"]

ticks=np.arange(len(x_axis))
plt.xticks(ticks, x_axis, rotation=75)
plt.hist(x_axis, bins=10)
plt.show
```

```python
#create a plots
# set x & y axis
x_axis=combined_grp_df["title"]
y_axis=combined_grp_df["salary"]

ticks=np.arange(len(x_axis))

plt.bar(x_axis, y_axis, align="center", alpha=0.5, color=["k", "r", "g", "m", "b", "c", "y"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation=75)

#set title
plt.title("Average salaries by employee title")
plt.xlabel("Employee Title")
plt.ylabel("Employee Salary")
# save plot
plt.savefig(".\images\salary_by_titles.png")
plt.show()
```
